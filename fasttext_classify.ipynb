```python
import fasttext
```

```python
import csv
from dataset_prepare.smt_preprocess import get_assertions
```

```python
model = fasttext.train_supervised(input="/Users/zhengyanglumacmini/Desktop/Projects/lm_smt/bin_sage2/z3_fasttext_train.txt")
```

```python
model.save_model("model_fasttext_z3.bin")
```

```python
test_dataset = []
with open("/Users/zhengyanglumacmini/Desktop/Projects/lm_smt/bin_sage2/raw/merged_3_mini_test.csv") as f:
    reader = csv.reader(f)
    next(reader)
    for row in reader:
        test_dataset.append((row[0], row[1]))
```

```python
test_dataset[:3]
```

```python
def ft_model_predict(model, path):
    assertions = get_assertions(path)
    predict = model.predict(assertions)
    return predict
```

```python
predictions = []
labels = []
correct = 0
total = 0
for path, label in test_dataset:
    predict = ft_model_predict(model, path)
    if predict[0][0] == "__label__yes":
        prediction = 1
    else:
        prediction = 0
    if label == "True":
        lb = 1
    else:
        lb = 0
    predictions.append(prediction)
    labels.append(lb)
    if prediction == lb:
        correct += 1
    total += 1
    print(f"correct/total: {correct}/{total}")
```

```python
with open("z3_fasttext_preds.csv", "w") as f:
    writer = csv.writer(f)
    writer.writerow(["preds", "labels"])
    for prediction, label in zip(predictions, labels):
        writer.writerow([prediction, label])
```
